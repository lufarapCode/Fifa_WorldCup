```python
import pandas as pd
```

### Leyendo los archivos cvs

```python
#Limpieza de datos
df_data_historica = pd.read_csv('fifa_worldCup_historical.csv')
df_data_fixture = pd.read_csv('fifa_fixture_2022_.csv')
df_data_missing = pd.read_csv('fifa_worldcup_missing_data.csv')
```

### Limpiando la data  fixture

```python
# Limpiar espacios en blanco que haya en una cadena de texto
df_data_fixture['home'] = df_data_fixture['home'].str.strip()
df_data_fixture['away'] = df_data_fixture['away'].str.strip()
```

### Limpiando data faltante y combniar a la data historia

```python
# comprobamos data faltante en una columna
# df_data_missing[df_data_missing['home'].isnull()]
# df_data_missing[df_data_missing['away'].isnull()]


```

```python
# eliminado esa data faltante
df_data_missing.dropna(inplace=True)
```

```python
 df_data_missing[df_data_missing['home'].isnull()]
```

```python
# combinando la data, el ignore_index nos ayuda a que los indices de las tablas se omitan y todo empiece desde 0
# ya que si no coloco eso, por ejemplo la data hiostoria del 1 al 100 y la data missing del 1 al 200, en la tabla general 
# iria del 1... 100, 1... 200 lo cual seria incorrecto
df_data_historica_merge = pd.concat([df_data_historica, df_data_missing], ignore_index=True)
df_data_historica_merge
```

```python
duplicados = df_data_historica_merge.duplicated()
print(duplicados)
```

```python
df_data_historica_merge.sort_values('year', inplace=True)
df_data_historica_merge
# Eliminar duplicados 
# df_data_historica_merge.drop_duplicates(inplace=True)
```

```python
# Eliminar duplicados 
df_data_historica_merge.drop_duplicates(inplace=True)
df_data_historica_merge
```
