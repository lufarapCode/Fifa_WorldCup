```python
# Expresiones regulares, con este caracter ^ -> busca lo que NO contenga
# \d  -> digitos
# - –  -> va buscar tambien ese caracter -
# en general con la expresion regular [^\d-] -> Buscame las filas que no contenga digitos (\d) ni este signo (-)
df_data_historica[df_data_historica['score'].str.contains('[^\d–]')]
```
